## Support Vector Classifier Implementation

```python
## importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
## Creating a synthetic data points
from sklearn.datasets import make_classification
X, y = make_classification( n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
## performing Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.25)
```

```python
## importing Support Vector Classifier
from sklearn.svm import SVC
```

Now here we need to decide the "Kernel" to be used. Different kernels give different results with the data, so its important to choose a good kernel for the dataset.  

```python
## Kernel - 'rbf'
"""
(Radial Basis Function) is flexible and works well in most cases even if the data is not linearly separable.
"""
svc = SVC(kernel='rbf') 
```

```python
## fitting the model
svc.fit(X_train, y_train)
```

```python
## Prediction
y_pred = svc.predict(X_test)
```

```python
## performance metrices
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)

print(c_report)
print(confusion_mat)
```

```python
## Kernel - 'poly' -> used mostly when the data has polynomial relationships
poly_kernel = SVC(kernel='poly')
poly_kernel.fit(X_train, y_train)
## Prediction
y_pred2 = poly_kernel.predict(X_test)

## performance metrices
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred2)
confusion_mat = confusion_matrix(y_test, y_pred2)

print(c_report)
print(confusion_mat)
```

```python
## Kernel - 'linear' -> used mostly when there is linear relationship between the data
linear_kernel = SVC(kernel='linear')
linear_kernel.fit(X_train, y_train)
## Prediction
y_pred3 = poly_kernel.predict(X_test)

## performance metrices
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred3)
confusion_mat = confusion_matrix(y_test, y_pred3)

print(c_report)
print(confusion_mat)
```

```python
## Kernel - 'sigmoid' -> very less used mostly in neural network activation
sigmoid_kernel = SVC(kernel='sigmoid')
sigmoid_kernel.fit(X_train, y_train)
## Prediction
y_pred4 = poly_kernel.predict(X_test)

## performance metrices
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred4)
confusion_mat = confusion_matrix(y_test, y_pred4)

print(c_report)
print(confusion_mat)
```

### Hyperparameter Tuning the model

#### 1. Using Grid Search Cross Validation

```python
from sklearn.model_selection import GridSearchCV

## defining the parameter range
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, refit = True, cv=5, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred5 = grid.predict(X_test)

## performance metrices
c_report = classification_report(y_test, y_pred5)
confusion_mat = confusion_matrix(y_test, y_pred5)

print(c_report)
print(confusion_mat)
```

#### 2. Using Randomized K Folds Cross Validation

```python
from sklearn.model_selection import RandomizedSearchCV

## defining parameters distributions
param_distributions = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
}
```

```python
random_search = RandomizedSearchCV(SVC(), param_distributions=param_distributions, cv=5, verbose=3)
random_search.fit(X_train, y_train)
```
